# Food price watch in Nigeria

### Brief description of Dataset

The Selected food price data represents different prices of food items in Lagos state and their comparison to different states in the country. It displayed the prices of different food items from January 2017 till August 2018. The data representation also displays the percentage increase/decrease in the price of food items from the year 2017 to 2018. It also reflected the percentage increase/decrease in the price of food items between January and August 2018. The remaining four columns displays the state with the highest price and the state with the lowest price of a particular food item respectively.
